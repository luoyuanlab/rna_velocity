```python
import os
os.chdir('..')
import scvelo as scv
import scanpy as sc
import cell2fate as c2f
import pickle as pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch
```

## Gastrulation erythroid

```python
adata = sc.read_h5ad('../Data/gastrulation_erythroid.h5ad')

adata = c2f.utils.get_training_data(adata, cells_per_cluster = 10**5, cluster_column = 'celltype', remove_clusters = [], min_shared_counts = 20, n_var_genes= 3000)

c2f.Cell2fate_DynamicalModel.setup_anndata(adata, spliced_label='spliced', unspliced_label='unspliced', batch_key = "sequencing.batch")

Tmax_prior_mean = 50
Tmax_prior_sd = 50
n_modules = c2f.utils.get_max_modules(adata)
mod = c2f.Cell2fate_DynamicalModel(adata, n_modules = n_modules, Tmax_prior={"mean": Tmax_prior_mean, "sd": Tmax_prior_sd})
mod.train()
adata = mod.export_posterior(adata)
mod.compute_and_plot_total_velocity(adata, save = False, delete = False, plot = False)

#scv.tl.velocity_graph(adata, vkey='Velocity', n_jobs=-1)
#scv.tl.velocity_embedding(adata, basis='umap', vkey='Velocity')
fix, ax = plt.subplots(1, 1, figsize = (8, 6))
scv.pl.velocity_embedding_stream(adata, basis='umap', save = False, vkey='Velocity', color='celltype', title='cell2fate', fontsize=20, show = False, ax = ax, legend_fontsize = 15)
plt.savefig('../Gastrulation_Erythroid_Results/cell2fate_UMAP_stream.png', dpi=80)
adata.write_h5ad('../Gastrulation_Erythroid_Results/cell2fate.h5ad')
```

## Simulated data

```python
adata = sc.read_h5ad('../Data/simulated_T_cell.h5ad')
adata.obsm['X_umap'] = pickle.load(open('../Data/UMAP.pkl','rb'))

adata = c2f.utils.get_training_data(adata, cells_per_cluster = 10**5, cluster_column = 'dataset', remove_clusters = [], min_shared_counts = 20, n_var_genes= 3)

c2f.Cell2fate_DynamicalModel.setup_anndata(adata, spliced_label='spliced', unspliced_label='unspliced')

Tmax_prior_mean = 50
Tmax_prior_sd = 50
n_modules = 10
mod = c2f.Cell2fate_DynamicalModel(adata, n_modules = n_modules, Tmax_prior={"mean": Tmax_prior_mean, "sd": Tmax_prior_sd})
mod.train()
adata = mod.export_posterior(adata)
mod.compute_and_plot_total_velocity(adata, save = False, delete = False, plot = False)

#scv.tl.velocity_graph(adata, vkey='Velocity', n_jobs=-1)
#scv.tl.velocity_embedding(adata, basis='umap', vkey='Velocity')
fix, ax = plt.subplots(1, 1, figsize = (8, 6))
scv.pl.velocity_embedding_stream(adata, basis='umap', save = False, vkey='Velocity', color='time', title='cell2fate', fontsize=20, show = False, ax = ax, legend_fontsize = 15)
plt.savefig('../Simulation_Results/cell2fate_UMAP_stream.png', dpi=80)
adata.write_h5ad('../Simulation_Results/cell2fate.h5ad')
```
