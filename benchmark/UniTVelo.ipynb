```python
import scvelo as scv
import scanpy as sc
import pickle as pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import unitvelo as utv
```

## Gastrulation erythroid

```python
adata = sc.read_h5ad('../Data/gastrulation_erythroid.h5ad')
scv.pp.filter_and_normalize(adata, min_shared_counts=20, n_top_genes=3000)
scv.pp.moments(adata, n_pcs=30, n_neighbors=30)

velo = utv.config.Configuration()
velo.N_TOP_GENES = 3000
velo.R2_ADJUST = True
velo.IROOT = None
velo.FIT_OPTION = '1'
velo.GPU = -1
velo.vgenes = 'offset'
scv.pp.neighbors(adata)
adata = utv.run_model(adata, 'celltype', config_file=velo)

scv.tl.velocity_graph(adata, vkey='velocity', n_jobs=-1)
scv.tl.velocity_embedding(adata, basis='umap', vkey='velocity')
fix, ax = plt.subplots(1, 1, figsize = (8, 6))
scv.pl.velocity_embedding_stream(adata, basis='umap', save = False, vkey='velocity', color='celltype', title='UniTVelo', fontsize=20, show = False, ax = ax, legend_fontsize = 15)
plt.savefig('../Gastrulation_Erythroid_Results/UniTVelo_UMAP_stream.png', dpi=80)
adata.write_h5ad('../Gastrulation_Erythroid_Results/UniTVelo.h5ad')
```
