```python
import scvelo as scv
import scanpy as sc
import pickle as pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import celldancer as cd
import celldancer.cdplt as cdplt
from celldancer.cdplt import colormap
from celldancer.utilities import export_velocity_to_dynamo
import celldancer.utilities as cdutil
```

## Gastrulation erythroid

```python
import pytorch_lightning as pl
trainer = pl.Trainer(progress_bar_refresh_rate=0)
trainer = pl.Trainer(enable_progress_bar=False)

#adata = sc.read_h5ad('../Data/gastrulation_erythroid.h5ad')
adata = sc.read_h5ad('/share/fsmresfiles/lyw/Singlecell/RNAVelocity/scvelo_examples/gastrulation_erythroid/gastrulation_erythroid.h5ad')
scv.pp.filter_and_normalize(adata, min_shared_counts=20, n_top_genes=3000)
scv.pp.moments(adata, n_pcs=30, n_neighbors=30)

input_data = cdutil.adata_to_df_with_embed(adata,
                                  us_para=['Mu','Ms'],
                                  cell_type_para='celltype',
                                  embed_para='X_umap',
                                  save_path='../cellDancer_result/cell_type_u_s_GE.csv')
loss_df, cellDancer_df=cd.velocity(input_data,
                                   gene_list= np.array(input_data['gene_name']),
                                   permutation_ratio=0.125,
                                   n_jobs=1)

cellDancer_df=cd.compute_cell_velocity(cellDancer_df=cellDancer_df, projection_neighbor_choice='gene', 
                                        expression_scale='power10', projection_neighbor_size=10, speed_up=(100,100))
adata = export_velocity_to_dynamo(cellDancer_df, adata)

adata.layers['velocity_S'] = adata.layers['velocity_S'].A
scv.tl.velocity_graph(adata, vkey='velocity_S', n_jobs=-1)
scv.tl.velocity_embedding(adata, basis='umap', vkey='velocity_S')
#fix, ax = plt.subplots(1, 1, figsize = (8, 6))
#scv.pl.velocity_embedding_stream(adata1, basis='umap', save = False, vkey='velocity_S', color='celltype', title='scVelo_dynamical', fontsize=20, show = False, ax = ax, legend_fontsize = 15)
#plt.savefig('../Gastrulation_Erythroid_Results/cellDancer_UMAP_stream.png', dpi=80)
#adata.write_h5ad('../Gastrulation_Erythroid_Results/cellDancer.h5ad')
```
