```python
import anndata as ad
import numpy as np
import scvelo as scv
import scanpy as sc
import torch
import deepvelo as dv
import pickle as pickle
import matplotlib.pyplot as plt
import pandas as pd
import time
```

## Gastrulation erythroid

```python
adata = sc.read_h5ad('../Data/gastrulation_erythroid.h5ad')
scv.pp.filter_and_normalize(adata, min_shared_counts=20, n_top_genes=3000)
scv.pp.moments(adata, n_pcs=30, n_neighbors=30)

param = dv.Constants.default_configs
param['n_gpu'] = 1
trainer = dv.train(adata, param)

scv.tl.velocity_graph(adata, vkey = 'velocity', n_jobs=-1)
scv.tl.velocity_embedding(adata, vkey = 'velocity')
fix, ax = plt.subplots(1, 1, figsize = (8, 6))
scv.pl.velocity_embedding_stream(adata, basis='umap', save = False, vkey='velocity', color='celltype', title='DeepVelo', fontsize=20, show = False, ax = ax, legend_fontsize = 15)
plt.savefig('../Gastrulation_Erythroid_Results/DeepVelo_UMAP_stream.png', dpi=80)
adata.write_h5ad('../Gastrulation_Erythroid_Results/DeepVelo.h5ad')
```
